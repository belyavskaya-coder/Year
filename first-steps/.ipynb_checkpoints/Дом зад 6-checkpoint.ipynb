```python
import re

# --- 1. Загружаем purchase_log.txt и строим словарь user_id → category ---
purchase_map = {}

print("Читаю purchase_log.txt и строю словарь категорий...")

with open('purchase_log.txt', 'r', encoding='utf-8') as f:
    content = f.read()

# Регулярное выражение: ищем {"user_id": "...", "category": "..."}
# Поддерживает пробелы, кавычки — по стандарту JSON
pattern = r'\{"user_id"\s*:\s*"([^"]+)"\s*,\s*"category"\s*:\s*"([^"]+)"\}'
matches = re.findall(pattern, content)

for uid, cat in matches:
    purchase_map[uid] = cat

print(f"Загружено {len(purchase_map)} записей из purchase_log.txt")

# --- 2. Обрабатываем visit_log__1_.txt построчно и формируем funnel.csv ---
print("Обрабатываю visit_log__1_.txt и формирую funnel.csv...")

count = 0

with open('visit_log__1_.txt', 'r', encoding='utf-8') as fin, \
     open('funnel.csv', 'w', encoding='utf-8') as fout:

    # Записываем заголовок
    fout.write('user_id,source,category\n')
    
    # Пропускаем первую строку, если там заголовок (как в примере: "user_id,source...")
    first_line = fin.readline().strip()
    if first_line.startswith('user_id'):
        # Это заголовок — продолжаем читать дальше
        pass
    else:
        # Это данные — вернём их в поток
        from io import StringIO
        fin = StringIO(first_line + '\n' + fin.read())

    for line in fin:
        line = line.strip()
        if not line:
            continue

        parts = line.split(',')
        if len(parts) < 2:
            continue

        user_id = parts[0]
        source = parts[1]

        # Есть ли событие типа email*?
        has_purchase = any(part.startswith('email') for part in parts[2:])
        
        if has_purchase and user_id in purchase_map:
            category = purchase_map[user_id]
            # Пропускаем записи с "не определена", если нужно — закомментируйте условие ниже
            if category == "не определена":
                continue
            fout.write(f'{user_id},{source},{category}\n')
            count += 1

print(f"✅ Готово! В funnel.csv записано {count} визитов с покупками.")
```
