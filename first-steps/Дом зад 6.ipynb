```python
import re

purchase_map = {}

with open('purchase_log.txt', 'r', encoding='utf-8') as f:
    content = f.read()

pattern = r'\{"user_id"\s*:\s*"([^"]+)"\s*,\s*"category"\s*:\s*"([^"]+)"\}'
matches = re.findall(pattern, content)

for uid, cat in matches:
    purchase_map[uid] = cat

count = 0

with open('visit_log__1_.txt', 'r', encoding='utf-8') as fin, \
     open('funnel.csv', 'w', encoding='utf-8') as fout:

    fout.write('user_id,source,category\n')

    for line in fin:
        parts = line.strip().split(',')
        if len(parts) < 2:
            continue
        user_id = parts[0]
        source = parts[1]
        has_purchase = any(part.startswith('email') for part in parts[2:])
        if has_purchase and user_id in purchase_map:
            category = purchase_map[user_id]
            fout.write(f'{user_id},{source},{category}\n')
            count += 1

print(f"✅ Готово! Записано {count} строк в funnel.csv")
```
